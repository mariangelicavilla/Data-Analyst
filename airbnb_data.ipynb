2. A basic GROUPED bar chart comparing columns neighbourhood_group and number_of_reviews

```python
# Assuming 'neighbourhood_group' and 'number_of_reviews' are columns in your DataFrame
grouped_data = data.groupby('neighbourhood_group')['number_of_reviews'].mean()

# Plotting a grouped bar chart
grouped_data.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Mean Number of Reviews')
plt.title('Mean Number of Reviews by Neighbourhood Group')

# Display the plot
plt.show()
```

3. A basic STACKED and Pivot bar chart

```python
# Assuming 'neighbourhood_group' and 'number_of_reviews' are columns in your DataFrame
pivot_data = data.pivot_table(index='neighbourhood_group', values='number_of_reviews', aggfunc='sum')

# Plotting a stacked bar chart
pivot_data.plot(kind='bar', stacked=True)

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Number of Reviews')
plt.title('Total Number of Reviews by Neighbourhood Group')

# Display the plot
plt.show()
```

5. Boxplot

Comparing prices during 2021 in three different neighbourhood groups

```python
# Assuming 'data' is your DataFrame
data_copy = data.copy()  # Create a copy of the original DataFrame

# Convert 'last_review' to datetime on the copied DataFrame
data_copy['last_review'] = pd.to_datetime(data_copy['last_review'])

# Filter data for the year 2021 and specific neighborhood groups on the copied DataFrame
selected_year = 2021
selected_neighborhoods = ['Centro', 'Latina', 'Retiro']
filtered_data = data_copy[
    (data_copy['last_review'].dt.year == selected_year) & 
    (data_copy['neighbourhood_group'].isin(selected_neighborhoods))]

# Create a boxplot to compare 'price' across 'neighbourhood_group'
plt.figure(figsize=(14, 8))
sb.boxplot(x='neighbourhood_group', y='price', data=filtered_data)
plt.title(f'Comparison of Prices Across Selected Neighbourhood Groups for {selected_year}')
plt.show()
```
